```python
def scrape_to_csv():
    url = 'https://www.valuepenguin.com/average-cost-of-wedding'
    response = requests.post(url=url)
    doc = lh.fromstring(response.content)
    tr_elements = doc.xpath('//tr')

    col = []
    for i, t in enumerate(tr_elements[0].iterchildren()):
        n = i + 1
        data = t.text_content()
        col.append((data.strip(), []))

    for j in range(1, len(tr_elements)):
        row = tr_elements[j]
        if len(row) != n:
            continue

        for i, t in enumerate(row.iterchildren()):
            data = t.text_content().strip()
            if i > 0:
                try:
                    data = int((sub(r'[^\d.]', '', data)))
                except Exception:
                    data = None
            col[i][1].append(data)

    _dict = {title: column for (title, column) in col}
    df = pd.DataFrame(_dict)

    data = df.set_index('State')
    data.to_csv('rob.csv')
    return df
```
